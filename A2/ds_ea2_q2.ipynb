```python
import pandas as pd
import numpy as np
import math
```

```python
countries = pd.read_csv("https://github.com/WHPAN0108/BHT-DataScience-S23/raw/main/python-DS/country.csv", delimiter=',')
countries
```

```python
countries.describe()
```

```python
pd.DataFrame(countries.values[-4:], columns=countries.columns)
```

```python
countries.where(lambda record: record["Currency"] == "EUR").dropna()
```

```python
currencyCountries = pd.DataFrame(countries[["Name", "Currency"]])
currencyCountries
```

```python
countries.where(lambda record: record["GDP"] > 2000).dropna()
```

```python
countries.where(lambda record: 50000000 < record["Population"]).where(lambda record: record["Population"] < 150000000).dropna()
```

```python
np.average(countries.dropna()["GDP"])
```

```python
np.average(countries["GDP"].replace(math.nan, 0))
```

```python
countries["Density"] = list(map(lambda record: record[1] / record[2], countries.values))
countries
```

```python
countries.sort_values("Name")
```

```python
newAreaCountries = pd.DataFrame(countries)
newAreaCountries['Area'] = newAreaCountries['Area'].apply(lambda record: "SMALL" if record <= 1000000 else "BIG")
newAreaCountries
```
